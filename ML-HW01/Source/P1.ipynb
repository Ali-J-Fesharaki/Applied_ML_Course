```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data_set = pd.read_csv('./Dataset_01.csv',index_col='index')
data_set.drop(columns='Unnamed: 0',inplace=True)#removing the unnamed column for redunanacy of information(similar to index column)
data_set.drop(columns='edu',inplace=True)#removing the job column for redunanacy of information(similar to edu_level column)
data_set.drop(columns='gain',inplace=True)
data_set.drop(columns='loss',inplace=True)
data_set.drop(columns='nationality',inplace=True)
data_set.drop(columns='race',inplace=True)
data_set.head()
data_set=data_set.iloc[0:]
data_set.count()
```

```python
data_set["hours-per-week"].value_counts(normalize=True)
```
